```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import joblib
```

```python
df=joblib.load('dfo.joblib')
df
```

```python
fig, axes = plt.subplots(2,3, figsize=(10,6))
axes[0,0].scatter(df.lon, df.lat, alpha = 0.1,s=.2)
axes[0,1].scatter(df.year,df.lon, alpha =0.1,s=.2)
axes[0,2].scatter(df.year,df.lat, alpha =0.1,s=.2)
axes[1,0].hist(df.lat,30, alpha =0.1)
axes[1,1].hist(df.carea, 20, log=True, alpha =0.1)
axes[1,2].scatter(df.year,df.carea, alpha =0.1,s=.2)
axes[0,0].set_title("lon-lat")
axes[0,1].set_title("year-lon")
axes[0,2].set_title("year-lat")
axes[1,0].set_title("lat")
axes[1,1].set_title("carea")
axes[1,2].set_title("year-carea")


```
