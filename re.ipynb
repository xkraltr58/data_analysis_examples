```python
import pandas as pd
import numpy as np
import re

df=pd.read_csv("/content/strdata.txt" ,sep="\s+")
```

```python
df["Fullname"] = df["Name"].astype(str) + " "+ df["Surname"].astype(str)
df
```

Aufgabe 2)

```python
df["Country_codes"]=df["Country"].astype('category').array.codes
df["Country_codes"]= df["Country_codes"].replace([1,0,2],[90,49,44])
df
```

aufgabe 3)

```python
arr1=[]
arr2=[]

for i in df.Tel:
  arr1.append(str(i)[:3])
  arr2.append(str(i)[3:9])
  
df["area_code"]= arr1
df
```

aufgabe 4)

```python
arr0=[]
arr1=[]
arr2=[]

pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)

for i in df["e-mail"]:
  reg=regex.match(str(i))
  arr0.append(reg.groups()[0])
  arr1.append(reg.groups()[1])
  arr2.append(reg.groups()[2])

df['username']= arr0
df['domain']=arr1
df['suffix']=arr2
df
```
